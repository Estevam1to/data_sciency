```python
import pickle
from sklearn.ensemble import RandomForestClassifier
```

```python
with open('../database/census.pkl', 'rb') as f:
    x_census_treinament, y_census_treinament, x_census_test, y_census_test = pickle.load(f)
```

```python
x_census_treinament.shape, y_census_treinament.shape
```

```python
random_forest_census = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
random_forest_census.fit(x_census_treinament, y_census_treinament)
```

```python
previsoes = random_forest_census.predict(x_census_test)
previsoes
```

```python
y_census_test
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_census_test, previsoes)
```

```python
print(classification_report(y_census_test, previsoes))
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(random_forest_census)
cm.fit(x_census_treinament, y_census_treinament)
cm.score(x_census_test, y_census_test)
```
